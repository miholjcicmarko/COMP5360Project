```python
import pandas as pd
```

```python
data = pd.read_csv("/Users/markomiholjcic/Documents/GitHub/COMP5360Project/project_files/csv_files/CLEANCDRData.csv")
data
```

```python
from datetime import datetime
groupDate = data.groupby("Last Update")
df1 = groupDate.count()

df1 = list(df1.index.values)
df1.sort(key = lambda date: datetime.strptime(date, '%m/%d/%y'))

dfDate = pd.DataFrame(df1)
dfDate
```

```python
dataChina = data[data["Country/Region"] == "China"]  
dataChina
groupChina = dataChina.groupby("Last Update")
dChina = groupChina.sum()
dChina = list(dChina["Confirmed"].values)
dChina.sort()
dChina

dfChina = pd.DataFrame({"China": dChina})
dfChina = pd.concat([dfDate,dfChina], axis=1, sort = False)
dfChina = dfChina.rename(columns={0: "Date"})
dfChina = dfChina.set_index('Date')
dfChina[10:]
```

```python
dataUS = data[data["Country/Region"] == "US"]  
groupDate = dataUS.groupby("Last Update")
df1 = groupDate.count()
df1

df1 = list(df1.index.values)
df1.sort(key = lambda date: datetime.strptime(date, '%m/%d/%y'))

dfDate = pd.DataFrame(df1)
dfDate
```

```python
dataUS = data[data["Country/Region"] == "US"]  
dataUS
groupUS = dataUS.groupby("Last Update")
dUS = groupUS.sum()
dUS = list(dUS["Confirmed"].values)
dUS.sort()
dUS

dfUS = pd.DataFrame({"US": dUS})
dfUS = pd.concat([dfDate,dfUS], axis=1, sort = False)
dfUS = dfUS.rename(columns={0: "Date"})
dfUS = dfUS.set_index('Date')
dfUS
```

```python
dfdata = pd.concat([dfChina,dfUS], axis=1, sort = False)
dfdata
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10, 6)

dfChina.plot(grid = True);
dfUS.plot(grid = True);
dfdata.plot(grid = True);
```
